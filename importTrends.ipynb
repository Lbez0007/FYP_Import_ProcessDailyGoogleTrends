# Import Libraries

```python
import pandas as pd
from pytrends.request import TrendReq
from pytrends import dailydata
from datetime import date
```

# Defining Parameters

```python
# Timezone Offset(US CST = 360)
pytrends = TrendReq(hl='en-US', tz=360)

keyword_list = ["pandemic","covid","covid-19","coronavirus","cases",
           "cdc","fauci","recession","deaths","restrictions",
           "lockdown","vaccine","vaccines","travel","stimulus"]
```

# Google Trends Retrieval

```python
def calculate_daily_trend(key):
    daily_trends = dailydata.get_daily_data(key, 2017, 1, 2020, 12, geo = 'US')
    return daily_trends

df = pd.DataFrame(index=pd.date_range(date(2017,1,1), date(2020,12,31), freq='D'))

for key in keyword_list:
    daily_trends = calculate_daily_trend(key)
    df[key] = 0
    for index, row in daily_trends.iterrows():
        df.at[index, key] = row[:1]

df.to_csv("GoogleTrends.csv", index=True)
```
